```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from random import seed
from random import random
import math
import sys

df = pd.read_csv("ghosts_ghouls_goblins.csv")

base_features = len(df.columns) - 3

#Establishes 1-hot encoding for the different color inputs
dummies = pd.get_dummies(df['color'])
df = df.drop('color', axis = 1)
df = df.join(dummies)
features  = base_features + len(dummies.columns)


#Establishes 1-hot encoding for the different type outputs
dummies = pd.get_dummies(df['type'])
df = df.drop('type', axis = 1)
df = df.join(dummies)
targets = len(dummies.columns)

def Get_Color(data):
    color = []
    for index in range(len(data)):
        if data['black'][index]:
            color.append('k')
        elif data['blue'][index]:
            color.append('b')
        elif data['blood'][index]:
            color.append('r')
        elif data['clear'][index]:
            color.append('y')
        elif data['green'][index]:
            color.append('g')
        elif data['white'][index]:
            color.append('m')
            
    return color;

def Plot(data, plot, columns, col_x,col_y,color):
    x = columns[col_x]
    y = columns[col_y]
    plot.set_xlabel(x)
    plot.set_ylabel(y)
    plot.set_title(x + " vs " + y)
    plot.scatter(data[x],data[y],c=color)

def Cross_Plot(data):
    cols = data.columns
    num_cols = base_features
    color = Get_Color(data)
    plt.figure(figsize=(60,40),facecolor='white')
    for col in range(1,num_cols):
            for cross_col in range(1,num_cols):
                if col != cross_col:
                    ax = plt.subplot2grid((num_cols, num_cols), (col, cross_col))
                    Plot(data, ax,cols,col, cross_col,color)
    
    plt.show()

Cross_Plot(df)

#Splits the training data into a train/test set and a validation set
train_test = df[:373]
validation = df[373:]

def Get_KFolds(data, y_label, num_folds):
    #Creates 5 folds from the train/test set each with a separate training and test set
    folds = []
    kf = KFold(n_splits = num_folds)
    tempdf = data.drop('id',axis = 1)
    for train_index, test_index in kf.split(data):
        training = []
        test = []
        
        train_x = tempdf.drop(y_label, axis=1).values
        train_y = tempdf[y_label].values
        
        #Creates a training set within the fold
        x = []
        y = []
        
        for index in train_index:
            x.append(train_x[index])
            y.append(train_y[index])
        training = [x,y]
        
        #Creates a test set within the fold
        x = []
        y = []
        for index in test_index:
            x.append(train_x[index])
            y.append(train_y[index])
        test = [x,y]

        folds.append([training,test])
    
    return folds

seed(1)
folds = Get_KFolds(train_test, ['Ghost', 'Ghoul', 'Goblin',], 5);
```
